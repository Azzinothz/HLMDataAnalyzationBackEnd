```python
origin_hlm = open("hlm.txt", "r").read()  # 读取《红楼梦》原文

print(origin_hlm[: 1000], "...")
```

```python
import re

preprocessing_hlm = re.sub("正文 第.{1,5}回", "", origin_hlm)  # 去除标题

# 去除需要忽略的字符
for char in "『』［］[]〔〕":
    preprocessing_hlm = "".join(preprocessing_hlm.split(char))

# 用“#”号分割句子
for char in " …《》，、。？！；：“”‘’'\n\r-=—()（）.【】":
    preprocessing_hlm = "#".join(preprocessing_hlm.split(char))

# 去除多余的“#”号
while "##" in preprocessing_hlm:
    preprocessing_hlm = "#".join(preprocessing_hlm.split("##"))
    
print(preprocessing_hlm[: 500], "...")
```

```python
chapters = re.sub("正文 第.{1,5}回", "$", origin_hlm).split("$")  # 再根据章节进行分割

# 再对每一章节进行上述的预处理
for chapter_number, chapter_text in enumerate(chapters):
    chapter_text = re.sub("正文 第.{1,5}回", "", chapter_text)  # 去除标题
    for char in "『』［］[]〔〕":
        chapter_text = "".join(chapter_text.split(char))
    for char in " …《》，、。？！；：“”‘’'\n\r-=—()（）.【】":
        chapter_text = "#".join(chapter_text.split(char))
    while "##" in preprocessing_hlm:
        chapter_text = "#".join(chapter_text.split("##"))
    chapters[chapter_number] = chapter_text

print("✏️已存储的章节总数：", len(chapters) - 1)
```

```python
import suffix_tree
print("开始构建后缀树...")
tree = suffix_tree.SuffixTree(preprocessing_hlm + "$")
tree.update_counter()
print("构建成功")
```

```python
print("开始为倒序文本构建后缀树 ...")
reversed_tree = suffix_tree.SuffixTree(preprocessing_hlm[::-1] + "$")
reversed_tree.update_counter()
print("构建成功")
```

```python
import dict_creat

# 进行句子计数
sentence_length_count = dict_creat.count_sentence_length(preprocessing_hlm)
count_of_length = [dict_creat.count_all_possibilities(sentence_length_count, i)
                    for i in range(len(sentence_length_count))]
print(count_of_length)
```
